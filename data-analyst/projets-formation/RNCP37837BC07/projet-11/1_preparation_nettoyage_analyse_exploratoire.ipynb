```python
#On affiche les doublons de 'Pays' dans le dataframe 'pays'
pays[pays.duplicated(subset=['Pays'], keep=False)]
```

```python
pays = pays[~(pays['Code Pays'] == 'F351')].reset_index(drop=True)
pays[pays.duplicated(subset=['Pays'], keep=False)]
```

```python
pays[pays.duplicated(subset=['Code Pays'], keep=False)]
```

```python
doublons_code_pays = ['Bolivie (État plurinational de)','Chine, continentale','Congo, République démocratique du','Congo, République du','Curacao','Îles Caïmans','Tchéquie','Égypte, République arabe d’','Micronésie (États fédérés de)','Royaume-Uni de Grande-Bretagne et d\'Irlande du Nord','Chine, RAS de Hong Kong','Iran (République islamique d\')','République kirghize','Corée, République de','Région administrative spéciale de Macao, Chine','Saint-Martin (fr)','République de Moldova','Îles Mariannes du Nord','Pays-Bas (Royaume des)','Puerto Rico (US)','Corée, République démocratique de','Cisjordanie et Gaza','République fédérale de Somalie','République slovaque','Sint Maarten (Dutch part)','Türkiye','République-Unie de Tanzanie','États-Unis d\'Amérique','Venezuela (République bolivarienne du)','Îles Vierges (EU)','Yémen, Rép. du']

pays = pays[~pays['Pays'].isin(doublons_code_pays)].reset_index(drop=True)

pays[pays.duplicated(subset=['Code Pays'], keep=False)].head()
```

```python
#Joindre les différents datasets sur la colonne 'Code Pays'
df = pd.merge(pays, politique, on=['Code Pays'], how='outer')
df = pd.merge(df, economique, on=['Code Pays'], how='outer')
df = pd.merge(df, social_1, on=['Code Pays'], how='outer')
df = pd.merge(df, social_2, on=['Code Pays'], how='outer')
df = pd.merge(df, technologie_1, on=['Code Pays'], how='outer')
df = pd.merge(df, technologie_2, on=['Code Pays'], how='outer')
df = pd.merge(df, technologie_3, on=['Code Pays'], how='outer')
df = pd.merge(df, environnement, on=['Code Pays'], how='outer')
```

```python
df = df.drop(['Code Pays'], axis=1)
df.head()
```

<div style="border: 1px solid RGB(0,150,250);" >
<h3 style="margin: auto; padding: 20px; color: RGB(0,150,250); ">3.2 - Analyse exploratoire</h3>
</div>

```python
#Afficher les dimensions du dataset
print("Le tableau comporte {} ligne(s)".format(df.shape[0]))
print("Le tableau comporte {} colonne(s)".format(df.shape[1]))
```

```python
#Vérifier s'il y a des valeurs manquantes
print(df.isnull().sum())
```

```python
df = df[~df['Pays'].isnull()]
df = df[~df['Stabilité politique, rang (%)'].isnull()]
df = df[~df['PIB par habitant'].isnull()]
df.loc[df['% du total des UGB'].isnull(), ['% du total des UGB', 'Prix de production (USD/tonne)']] = df.loc[df['% du total des UGB'].isnull(), ['% du total des UGB', 'Prix de production (USD/tonne)']].fillna(0)
df['Prix de production (USD/tonne)'] = df['Prix de production (USD/tonne)'].fillna(df['Prix de production (USD/tonne)'].mean())
df['Importation de poulet (tonne)'] = df['Importation de poulet (tonne)'].fillna(0)
df['Émission de CO2eq (kt)'] = df['Émission de CO2eq (kt)'].fillna(df['Émission de CO2eq (kt)'].mean())
df = df[~(df['Pays'] == 'Kosovo')]
print(df.isnull().sum())
```
