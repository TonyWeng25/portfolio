```python
#On trie les clients par chiffre d'affaires
df_clients = df_clients.sort_values(by='price')
df_clients.head()
```

```python
#On va calculer les proportions cumulées
df_clients['cumulative'] = df_clients['price'].cumsum() / df_clients['price'].sum()
df_clients['client_pct'] = np.arange(1, len(df_clients) + 1) / len(df_clients)
```

```python
#On crée df_lorenz en ajouter le point (0,0)
df_lorenz = pd.concat([pd.DataFrame({'client_pct': [0], 'cumulative': [0]}), df_clients[['client_pct', 'cumulative']]], ignore_index=True)
```

```python
#On calcule l'indice de Gini
gini_index = 1 - 2 * np.trapezoid(df_lorenz['cumulative'], df_lorenz['client_pct'])
print(f"Indice de Gini : {gini_index:.4f}")
```

```python
#On trace la courbe de Lorenz
fig = px.line(df_lorenz, x='client_pct', y='cumulative',
              title='Courbe de Lorenz du chiffre d\'affaires par client',
              labels={'client_pct': 'Proportion cumulée des clients', 'cumulative': 'Proportion cumulée du chiffre d\'affaires'})
fig.update_layout(title_x=0.5, title_y=0.95)

#On ajoute la ligne d'égalité parfaite
fig.add_scatter(x=[0, 1], y=[0, 1], mode='lines', line=dict(dash='dash'), name='Égalité parfaite')

#On ajoute l'indice de Gini
fig.add_annotation(text=f"Indice de Gini : {gini_index:.4f}", xref='paper', yref='paper', x=0.5, y=1.1, showarrow=False)

#On ajoute les points sur la courbe de Lorenz qui passe en x=0.5 ou y=0.5
fig.add_scatter(x=[0.5], y=[df_lorenz[df_lorenz['client_pct'] >= 0.5]['cumulative'].iloc[0]], mode='markers', name='50% des clients', marker=dict(color='yellow', size=10))
fig.add_scatter(x=[df_lorenz[df_lorenz['cumulative'] >= 0.5]['client_pct'].iloc[0]], y=[0.5], mode='markers', name='50% du CA', marker=dict(color='lightgreen', size=10))

fig.show()
```

```python
from scipy.stats import chi2_contingency
```

H0 : La catégorie d'un livre acheté est indépendante du genre du client

H1 : La catégorie d'un livre acheté dépend du genre du client 

```python
# On visualise la relation entre le genre d'un client et les catégories des livres achetés avec une heatmap
fig = px.density_heatmap(
    df_non_b2b,
    x='sex',
    y='categ',
    histfunc='count',
    title="Répartition des catégories de livres achetés par genre (hors BtoB)",
    color_continuous_scale='Blues'
)
#Changer le nom du count
fig.update_layout(xaxis_title='Genre du client', yaxis_title='Catégorie du livre')
fig.update_layout(title_x=0.5, title_y=0.95)
fig.update_layout(coloraxis_colorbar_title='Nombre de livres achetés')
fig.update_traces(texttemplate='%{z}', textfont_size=12)
fig.update_layout(width=600)
fig.show()
```

```python
#On crée un tableau de contingence pour le genre d'un client et les catégories des livres achetés
contingency_table = pd.crosstab(df_non_b2b['sex'],df_non_b2b['categ'])
print("Tableau de contingence :\n", contingency_table)
```

```python
#On effectue le test du Chi2
chi2, p, dof, expected = chi2_contingency(contingency_table)

#On affiche les résultats du test
print(f"Chi2: {chi2}, p-value: {p}, dof: {dof}")
print("Tableau des fréquences attendues:")
print(expected)
```

Hors clients BtoB, on observe une valeur du Chi-2 de 22.7, ce qui représente l'écart entre la table de contingence étudiée et la table des fréquences attendues si les deux variables sont indépendantes.

On observe un degré de liberté de 2, qui suppose que le test est relativement fiable.

Enfin, la p-value de 1.20*10^-5, inférieure au seuil de 0.05, nous confirme qu'on peut rejeter H0, donc qu'il y a une association significative entre le genre d'un client et la catégorie des livres achetés.

```python
chi2, p, dof, expected = chi2_contingency(contingency_table)

#On affiche les résultats du test
print(f"Chi2: {chi2}, p-value: {p}, dof: {dof}")
print("Tableau des fréquences attendues:")
print(expected)
```

H0 : Au moins une moyenne d'âge est différente.

```python
#Boxplot de l'âge des clients par catégorie de produits
fig = px.box(df_non_b2b, x='categ', y='age', 
             title='Boxplot de l\'âge des clients par catégorie de produits (hors BtoB)',
             labels={'categ': 'Catégorie de produits', 'age': 'Âge du client'},
             color='categ', color_discrete_map={'0': '#EF553B', '1': '#636EFA', '2': '#00CC96'})
fig.update_layout(xaxis_title='Catégorie de produits', yaxis_title='Âge du client',
                  title_x=0.5, title_y=0.95)
fig.update_traces(boxpoints=False)
fig.update_layout(width=800)
fig.show()
```

```python
categ_0 = df_non_b2b[df_non_b2b['categ'] == '0']['age']
categ_1 = df_non_b2b[df_non_b2b['categ'] == '1']['age']
categ_2 = df_non_b2b[df_non_b2b['categ'] == '2']['age']

print("Moyenne de l'âge pour la catégorie 0 : ", categ_0.mean())
print("Moyenne de l'âge pour la catégorie 1 : ", categ_1.mean())
print("Moyenne de l'âge pour la catégorie 2 : ", categ_2.mean())
```
